```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('robot.png')
img2=cv2.imread('robot2.png')
img=img[50:50+399,:,:]
```

```python
img.shape,img2.shape
```

```python
def showimg(im):
    plt.imshow(im[:,:,::-1])
    plt.show()
```

```python
def imshow(im):
    cv2.imshow('Preview',im)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
```

```python
showimg(img)
```

```python
showimg(img2)
```

```python
add1=cv2.add(img,img2)
showimg(add1)
```

```python
sub=cv2.subtract(img,img2)
showimg(sub)
```

```python
import numpy as np
```

```python
img2_1=cv2.multiply(img2,20*np.ones(img2.shape).astype('uint8'))
showimg(img2_1)
```

```python
img2_2=cv2.convertScaleAbs(img2,alpha=1.3,beta=50)
showimg(img2_2)
```

```python
img2_3=cv2.resize(img2,(400,400),interpolation=cv2.INTER_CUBIC)
img2_3.shape
```

```python
showimg(img2_3)
```

```python
img2_4=cv2.resize(img2,None,fx=10,fy=10,interpolation=cv2.INTER_CUBIC)
img2_4.shape
```

```python
showimg(img2_4)
```

```python
#img ROtation
from skimage.transform import rotate
image_5=rotate(img2,angle=180,resize=True)

showimg(image_5)
```

```python
from time import sleep
repeat=True
while repeat:
    for i in range(0,360,5):
        i=rotate(img2,angle=i,resize=True)
        cv2.imshow('Preview',i)
        key=cv2.waitKey(1)
        if key==ord('q'):
            repeat=False
            break
        sleep(0.1)
        
cv2.destroyAllWindows()
```
