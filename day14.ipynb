```python
import cv2,plotly,plotly.express as px,skimage
```

```python
def imshowPx(im,cv=True,gray=True):
    fig=px.imshow(im[:,:,::-1] if cv else im,binary_string=gray)
    fig.update_xaxes(visible=False)
    fig.update_yaxes(visible=False)
    plotly.io.show(fig)
```

```python
img=skimage.data.astronaut()
imshowPx(img,cv=False)
```

```python
blue_plane=img[:,:,-1].copy()#as this is not read from open cv hence already rgb 
gray_img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY) #RGB TO GRAY


blue_img=cv2.subtract(blue_plane,gray_img)
th,blue_binary=cv2.threshold(blue_img,55,255,cv2.THRESH_BINARY)


blue_binary2=skimage.morphology.remove_small_objects(blue_binary.astype(bool),134)
blue_binary3=skimage.morphology.remove_small_holes(blue_binary2.astype(bool),134)

strel=cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(10,10))
blue_binary4=cv2.dilate(blue_binary3.astype('uint8'),strel,iterations=1)
blue_binary5=skimage.morphology.remove_small_holes(blue_binary4.astype(bool),5000)

labels=skimage.measure.label(blue_binary5)
rp=skimage.measure.regionprops(labels,blue_binary5)
imshowPx(blue_binary5.astype('uint8'),gray=True,cv=False)
```

```python
img_orig=img[:,:,::-1].copy()
bbox=rp[0].bbox
cv2.rectangle(img_orig,pt1=(bbox[1],bbox[0]),pt2=(bbox[-1],bbox[-2]),color=(255,255,0),thickness=10)

cv2.imshow('',img_orig)
```
